```python
# Importing necessary libraries
from transformers import pipeline
```

```python
# Zero-shot classification
classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about motorcycle maintenance and repair.",
    candidate_labels=["education", "transportation", "business"],
)
```

```python
# Text generation
generator = pipeline("text-generation")
generator("In this course, we will teach you how to")

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "In this course, we will teach you how to",
    max_length=30,
    num_return_sequences=2,
)
```

```python
# Fill-mask
unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> motorcycles.", top_k=2)
```

```python
# Named entity recognition
ner = pipeline("ner", grouped_entities=True)
ner("My name is John and I work at a motorcycle dealership in Los Angeles.")
```

```python
# Question answering
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Motor adalah kendaraan bermotor yang memiliki dua roda dan biasanya digunakan untuk transportasi."
question = "Apa itu motor?"
```

```python
result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
# Sentiment analysis
classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a motorcycle repair workshop my whole life.")
```

```python
# Summarization
summarizer = pipeline("summarization")
summarizer(
    """
    The motorcycle industry has seen significant growth in recent years.
    Not only has the demand for motorcycles increased in urban areas due to
    traffic congestion, but manufacturers are also focusing on electric motorcycles
    to meet environmental standards. In countries like India and Vietnam,
    motorcycles are not just a mode of transport; they are a vital part of daily life.
    However, issues like road safety and maintenance are crucial for the sustainability of the industry.

    Many manufacturers are now investing in better safety features and rider training programs.
    As electric vehicles gain popularity, the motorcycle industry is also adapting to this trend,
    ensuring that it remains relevant in the evolving transportation landscape.
"""
)
```

```python
# Translation
translator = pipeline("translation_id_to_en", model="Helsinki-NLP/opus-mt-id-en")
text_to_translate = "Hari ini aku pergi ke bengkel motor"
result = translator(text_to_translate)
print(result[0]['translation_text'])
```

```python
# Question answering
qa_pipeline = pipeline("question-answering")
context = "Motor adalah kendaraan bermotor yang memiliki dua roda dan biasanya digunakan untuk transportasi."
question = "Apa itu motor?"
```

```python
result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

# **Analisis Tugas NLP Terkait Motor**

# **Klasifikasi Zero-Shot:**

**Tujuan:** Ini buat ngeliat jenis konten dari teks yang kita kasih tanpa perlu model khusus.

Implementasi: Kalimat "Ini adalah kursus tentang pemeliharaan dan perbaikan motor" diklasifikasikan ke label kayak "pendidikan," "transportasi," dan "bisnis."

Wawasan: Jadi, kursus ini punya relevansi di banyak bidang, keren kan? Bisa dibilang edukatif baik buat yang pengen belajar merawat motor atau mau buka usaha di bidang ini.

# **Generasi Teks:**
**Tujuan**: Buat nyiptain kelanjutan dari kalimat yang kita kasih.

**Implementasi:** Dengan prompt "Dalam kursus ini, kami akan mengajarkan Anda cara," model diminta buat bikin teks lanjutan.

**Wawasan:** Ini nunjukkin kalau modelnya bisa bikin konten yang informatif, kayak ngajarin skill penting buat merawat atau berkendara dengan motor.
Pengisian Masker (Fill-Mask):

**Tujuan:** Ini buat nebak kata yang pas buat ngisi tempat yang dimasker dalam kalimat.

**Implementasi:** Kalimat "Kursus ini akan mengajarkan kamu semua tentang <mask> motor." Model nyarankan istilah motor yang cocok.

**Wawasan:** Dari sini, kita bisa tahu topik-topik menarik yang bisa dibahas dalam edukasi motor. Model ini paham banget soal apa yang perlu dipelajari.

# **Pengenalan Entitas Bernama (NER):**

**Tujuan: **bold text** **Buat ambil dan kategorikan nama-nama penting dari teks.

**Implementasi:** Kalimat "Nama saya John dan saya kerja di dealer motor di Los Angeles" bikin model bisa nebak entitas penting seperti nama dan lokasi.

**Wawasan: **Ini nunjukkin hubungan personal dengan industri motor, menandakan pentingnya dealer lokal di komunitas.

# **Tanya Jawab:**

**Tujuan:** Buat kasih jawaban dari konteks yang udah ada.
Implementasi: Konteks tentang motor dipadukan sama pertanyaan "Apa itu motor?" dan model kasih jawaban yang relevan.

**Wawasan:** Model ini bisa kasih jawaban singkat dan akurat, jadi cocok banget buat alat edukasi atau jawaban untuk pertanyaan pelanggan seputar motor.
Analisis Sentimen:

**Tujuan:** Buat liat mood dari pernyataan yang ada.
